# **NumPy Problems**


**1**.**Basic Array Operations**

```python
import numpy as np
import pandas as pd

df = pd.read_csv("/content/1774404110.csv")
mpg = df["mpg"].to_numpy()
mean_mpg = np.mean(mpg)
median_mpg = np.median(mpg)
std_mpg = np.std(mpg)
count = np.sum(mpg > 25)
print(mean_mpg, median_mpg, std_mpg)
print("Count >25 mpg:", count)
```

**2.Filtering**

```python
cyl = df["cylinders"].to_numpy()
names = df["car name"].to_numpy()
filtered_names = names[cyl > 6]
print(filtered_names.tolist())
```

**3. Broadcasting**

```python
hp = df["horsepower"].replace("?", np.nan).astype(float).to_numpy()
mean_hp = np.nanmean(hp)
hp[np.isnan(hp)] = mean_hp
hp_new = hp * 1.10
print(hp_new)
```

**4. Boolean Indexing**

```python
disp = df["displacement"].to_numpy()
origin = df["origin"].to_numpy()

avg_disp = np.mean(disp[origin == 2])

print("Average displacement (Europe):", avg_disp)
```

**5. Conditional Aggregates**

```python
cyl = df["cylinders"].to_numpy()
mpg = df["mpg"].to_numpy()
unique_cyl = np.unique(cyl)
for c in unique_cyl:
    mean_val = np.mean(mpg[cyl == c])
    print(f"Cylinders {c}: Mean MPG = {mean_val}")
```

# **Pandas Problems**

**1. Basic Exploration**

```python
import pandas as pd
df = pd.read_csv("/content/1774404110.csv")
print(df.head(10))
print("Shape:", df.shape)
print(df.describe())
```

**2. Filtering and Indexing**

```python
filtered = df[(df["model year"] == 75) & (df["weight"] < 3000)]
result = filtered[["car name", "weight", "mpg"]]
print(result)
```

**3. Handling Missing Data**

```python
import pandas as pd
print(df.isnull().sum())
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
median_hp = df['horsepower'].median()
df['horsepower'] = df['horsepower'].fillna(median_hp)
```

**4. Data Transformation**

```python
df["power_to_weight_ratio"] = df["horsepower"] / df["weight"]
print(df[["horsepower", "weight", "power_to_weight_ratio"]].head())
```

**5. Group By**

```python
grouped = df.groupby("origin")["mpg"].mean()
print(grouped)
```

**6. Visualization Preparation**

```python
summary_df = df.groupby("model year")[["mpg", "weight", "horsepower"]].mean()
print(summary_df)
```

**7. Exporting Data**

```python
high_mpg = df[df["mpg"] > 30][["mpg", "cylinders", "horsepower", "weight"]]
high_mpg.to_csv("high_mpg_cars.csv", index=False)
print("File saved successfully!")
```
